```python
import pyfits as pf
import seaborn as sb
from pytransit import MandelAgol
from pytransit.orbits_f import orbits as of
from exotk.utils.orbits import as_from_rhop
warnings.simplefilter("ignore")
```

## Create the light curves

```python
tc, p, rho, b, k = 1, 2.5, 1.5, 0.5, 0.1
a = as_from_rhop(rho,p)
i = arccos(b/a)

filters = 'u g r i z J2 H2 Ks'.split()
npb  = len(filters)
ldcs = [[0.80,0.02],[0.61,0.16],[0.45,0.20],[0.36,0.20],
        [0.30,0.20],[0.21,0.22],[0.12,0.25],[0.12,0.19]]

white_noise_std = 0.75e-3

obs_duration = of.duration_circular(p,a,i) + 3./24.
obs_cadence  = 1./60./24.
obs_n = int(obs_duration/obs_cadence)

tm = MandelAgol()

time = linspace(1-0.5*obs_duration,1+0.5*obs_duration, obs_n)
nois_white = [normal(0,white_noise_std, size=obs_n) for ldc in ldcs]
flux_clean = [tm.evaluate(time, k, ldc, tc, p, a, i) for ldc in ldcs]
flux_obs = [flux+noise for flux,noise in zip(flux_clean, nois_white)]
```

## Export the data as FITS

```python
cols = ([pf.Column('time', 'D', 'BJD', array=time)] +
        [pf.Column('flux_{}'.format(filters[j]), 'D', array=flux_obs[j]) for j in range(npb)])
hdul = pf.HDUList([pf.PrimaryHDU(), pf.BinTableHDU.from_columns(cols)])
hdul[1].header['extname'] = 'flux_obs'
hdul.writeto('obs_data.fits', clobber=True)
```

## Plot the clean light curves

```python
with sb.axes_style('white'):
    fig,ax = subplots(1,1,figsize=(10,10))
    [ax.plot(time*24-24, flux-j*0.0025, '-') for j,flux in enumerate(flux_clean)]
setp(ax, ylim=(0.971,1.001), yticks=[])
fig.tight_layout()
```
