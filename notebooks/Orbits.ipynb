```python
from pytransit.orbits_f import orbits as of
```

```python
time = np.linspace(0,5, 5000)
t0, a, i, p, e, w = 2, 6, 0.5*pi, 5, 0.5, 0#.5*pi
```

```python
nthr = 1

ta_i = of.ta_eccentric(time, t0, p, e, w, nthr)
ta_n = of.ta_eccentric_newton(time, t0, p, e, w, nthr)
ta_s3 = of.ta_eccentric_s3(time, t0, p, e, w, nthr)
ta_s5 = of.ta_eccentric_series(time, t0, p, e, w, nthr)

z_i = of.z_eccentric(time, t0, p, a, i, e, w, nthr)
z_n = of.z_eccentric_newton(time, t0, p, a, i, e, w, nthr)
z_s3 = of.z_eccentric_s3(time, t0, p, a, i, e, w, nthr)
z_s5 = of.z_eccentric_s(time, t0, p, a, i, e, w, nthr)
```

```python
plot(ta_n)
plot(ta_i)
```

```python
fig,ax = subplots(2,1, figsize=(14,10))
ax[0].plot(time, cos(ta_i), 'k', lw=2)
ax[0].plot(time, 0.05+cos(ta_n), 'b--')
ax[0].plot(time, cos(ta_s5), 'k--')
ax[0].plot(time, cos(ta_s3), 'k:')
ax[1].plot(time, cos(ta_i)-cos(ta_s5), 'k')
ax[1].plot(time, cos(ta_i)-cos(ta_s3), 'k--')

fig.tight_layout()
```

```python
fig,ax = subplots(2,1, figsize=(14,10))
ax[0].plot(time, z_i, 'k', lw=2)
ax[0].plot(time, z_n, 'k', lw=2)
ax[0].plot(time, z_s5, 'k--')
ax[0].plot(time, z_s3, 'k:')
ax[1].plot(time, abs(z_n-z_s5), 'k')
ax[1].plot(time, abs(z_n-z_s3), 'k--')
fig.tight_layout()
```

```python
plot(time, z_n)
plot(time, z_s5)
xlim(0.9,2.1)
ylim(-0.1,2)
```
