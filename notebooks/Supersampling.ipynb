# Supersampling

```python
from pytransit.gimenez import Gimenez
from pytransit.orbits_f import orbits as of
```

```python
k, t0, p, a, i, e, w = 0.1, 1.01, 4, 8, 0.5*pi, 0.4, 0.5*pi
t_sc = np.linspace(0.9,1.1,200)
t    = np.linspace(0.9,1.1,7)

us = [0.4,0.1]
um = np.array([[0.0+0.08*iu, 0.0+0.025*i] for iu in range(10)])
```

```python
tm = Gimenez()
tm_ss = Gimenez(supersampling=4, exptime=t[1]-t[0])
```

```python
fl_sc = tm.evaluate(t_sc, k, us, t0, p, a, i, e, w)

fl = tm.evaluate(t, k, us, t0, p, a, i, e, w)
fl_ss = tm_ss.evaluate(t, k, us, t0, p, a, i, e, w)
```

```python
plot(t_sc, fl_sc, 'k:')
plot(t, fl, 'ko')
plot(t, fl_ss, 'k-', drawstyle='steps-mid')
ylim(0.987, 1.001)
```

```python
fig,ax = subplots(1,1,figsize=(14.8,5))
plot(tm_ss._time.reshape((tm_ss.npt,tm_ss.nss)).T, ones_like(tm_ss._time.reshape((tm_ss.npt,tm_ss.nss)).T), 'o')
plot(tm_ss.time, 0.01+ones_like(tm_ss.time), 'o')
ylim(0.99,1.02)
fig.tight_layout()
```
