```python
from pytransit.orbits_f import orbits as of
from pytransit.gimenez import Gimenez
```

```python
tm = Gimenez(npol=200)
time = np.linspace(0,4.5, 5000)
t0, a, i, p, e, w = 1.0, 5, 0.5*pi, 5, 0.3, 0.5*pi
```

```python
nthr = 4

z_i  = of.z_eccentric(time, t0, p, a, i, e, w, nthr)
z_ip = of.z_eccentric_ip(time, t0, p, a, i, e, w, nthr, True)
z_s3 = of.z_eccentric_s3(time, t0, p, a, i, e, w, nthr)
z_s5 = of.z_eccentric_s(time, t0, p, a, i, e, w, nthr)

f_i  = tm(z_i,  0.1, [0.2,0.1])
f_ip = tm(z_ip, 0.1, [0.2,0.1])
f_s3 = tm(z_s3, 0.1, [0.2,0.1])
f_s5 = tm(z_s5, 0.1, [0.2,0.1])
```

## Maximum error as a function of the argument of periastron

```python
wa = linspace(0,2*pi,200)
f_ia = array([tm(of.z_eccentric(time, t0, p, a, i, e, ww, nthr), 0.1, [0.2,0.1]) for ww in wa])
f_la = array([tm(of.z_eccentric_ip(time, t0, p, a, i, e, ww, nthr, True), 0.1, [0.2,0.1]) for ww in wa])
f_3a = array([tm(of.z_eccentric_s3(time, t0, p, a, i, e, ww, nthr), 0.1, [0.2,0.1]) for ww in wa])
f_5a = array([tm(of.z_eccentric_s(time, t0, p, a, i, e, ww, nthr), 0.1, [0.2,0.1]) for ww in wa])
```

```python
plot((f_la-f_ia).max(1))
plot((f_3a-f_ia).max(1))
plot((f_5a-f_ia).max(1), '--');
```

```python
fig,ax = subplots(2,1, figsize=(14,10))
ax[0].plot(time, f_i, 'k', lw=2)
ax[0].plot(time, f_s3, 'k--')
ax[0].plot(time, f_s5, 'k:')
ax[1].plot(time, 1e3*abs(f_ip-f_i), 'k-')
ax[1].plot(time, 1e3*abs(f_s3-f_i), 'k:')
ax[1].plot(time, 1e3*abs(f_s5-f_i), 'k--')
fig.tight_layout()
```
