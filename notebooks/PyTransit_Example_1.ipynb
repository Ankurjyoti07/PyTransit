# PyTransit example 1

This example shows the most straightforward way to calculate the transit model for a single and multiple sets of quadratic limb darkening coefficients.

## Preample

```python
import numpy as np
import matplotlib.pyplot as pl
from pytransit.gimenez import Gimenez
```

Initialise some sensible parameters for the transiting planet, create a time array `t`, and initialise the quadratic limb darkening coefficient arrays `us` and `um`. Here `us` contains a single set of quadratic limb darkening coefficients, while `um` contains 10 sets.

The parameters are:

 - `k `  planet to star radius ratio
 - `t0` zero epoch
 - `p`  orbital period
 - `a`  scaled semi-major axis
 - `i`  orbital inclination
 - `e`  orbital eccentricity
 - `w`  argument of periastron

```python
k, t0, p, a, i, e, w = 0.1, 1, 4, 8, 0.48*np.pi, 0.2, 0.5
t = np.linspace(0.9,1.1,500)

us = [0.4,0.1]
um = np.array([[0.0+0.08*iu, 0.0+0.025*i] for iu in range(10)])
```

## Model usage

Create an instance of the transit model, accept the default values

```python
m = Gimenez()
```

and evaluate the model for the two cases of limb darkening coefficient sets

```python
Is = m.evaluate(t, k, us, t0, p, a, i, e, w)
Im = m.evaluate(t, k, um, t0, p, a, i, e, w)
```

The `Gimenez.evaluate` method offers a direct way for the calculation of the light curve as a function of time `t`. It uses a circular 
orbit model if eccentricity is zero (or smaller than 1e-5), and slower eccentric orbit model with non-zero eccentricities.

The `Gimenez.__call__` method offers a way to evaluate the model as a function of normalised projected distance `z`. This is useful
if we don't want to recalculate the orbit but change the limb darkening coefficients or contamination factor, or if we have a better (faster)
method for the calculation of `z` than provided by the package.

Plot the light curves
---------------------

```python
pl.rc('font', size=14)
fig,ax = pl.subplots(1,2,figsize=(18,6))
ax[0].plot(t,Is, 'k');
ax[1].plot(t,Im + np.arange(um.shape[0])*0.001, 'k');
pl.setp(ax[0], ylim=(0.988,1.001), xlabel='Time [d]', ylabel='Normalised flux', title='Light curve for a single set of limb darkening coefficients');
pl.setp(ax[1], ylim=(0.988,1.011), xlabel='Time [d]', ylabel='Normalised flux', title='Light curve for a set of multiple limb darkening coefficients');
fig.tight_layout()
```
