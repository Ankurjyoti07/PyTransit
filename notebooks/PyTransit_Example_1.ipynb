PyTransit example 1
===================

Showing how to calculate the transit model for a single and multiple sets of quadratic limb darkening coefficients.

Preample
--------

Initialise some sensible parameters for the transiting planet, create a time array, and initialise the quadratic limb darkening coefficient arrays.

```python
import numpy as np
import matplotlib.pyplot as pl
from pytransit.orbits_f import orbits as of
from pytransit.gimenez import Gimenez
```

```python
k, t0, p, a, i, e, w = 0.1, 1, 4, 8, 0.48*np.pi, 0.2, 0.5
t = np.linspace(0.9,1.1,500)

us = [0.4,0.1]
um = np.array([[0.0+0.08*iu, 0.0+0.025*i] for iu in range(10)])
```

Model usage
-----------

Create an instance of the transit model class, accept the default values.

```python
m = Gimenez()
```

Calculate the normalised distances.

```python
z = of.z_eccentric(t, t0, p, a, i, e, w, nthreads=0)
```

...and evaluate the model.

```python
Is = m(z,k,us)
Im = m(z,k,um)
```

Plot the light curves
---------------------

```python
pl.rc('font', size=14)
fig,ax = pl.subplots(1,2,figsize=(18,6))
ax[0].plot(t,Is, 'k');
ax[1].plot(t,Im + np.arange(um.shape[0])*0.001, 'k');
pl.setp(ax[0], ylim=(0.988,1.001), xlabel='Time [d]', ylabel='Normalised flux', title='Light curve for a single set of limb darkening coefficients');
pl.setp(ax[1], ylim=(0.988,1.011), xlabel='Time [d]', ylabel='Normalised flux', title='Light curve for a set of multiple limb darkening coefficients');
fig.tight_layout()
```
